```python
import matplotlib.pylab as plt
import pandas as pd
```

```python
path="/Users/shriyaa/Desktop/CS598SS/project/real-case-runs-osprey-crash-correction/05/data_size595_k_10_run1"

f=open(path+'/track.txt','rb')
scores=[]

for line in f:
       	scores.append(float(line))

max_index=scores.index(max(scores))
print max_index

d={'data':scores}
df = pd.DataFrame(d)
df_avg=pd.rolling_mean(df,10)

vals=[]
for j in range(len(df_avg)):
       	vals.append(df_avg.data[j])
vals.append(df_avg.data[j])

plt.plot(vals,alpha=0.6,linewidth=3,c='gray',label='Rolling average')
plt.plot(scores,color='k')
plt.legend(frameon=False,loc=(0,0.85))
plt.ylabel('Score')
plt.xlabel('Iteration Number')
plt.ylim(5,15)
plt.xlim(0,400)
plt.plot(max_index-1,scores[max_index],'D')
```

```python
path="/Users/shriyaa/Desktop/CS598SS/project/real-case-runs-osprey-crash-correction/"

col=['r','g','b','k','m']

count=0
for item in ['01','02','03','04','05']:
    f=open(path+'/'+item+'/data_size595_k_10_run1/track.txt','rb')
    scores=[]
    for line in f:
        scores.append(float(line))
    max_index=scores.index(max(scores))
    plt.plot(scores,color=col[count],alpha=0.6)
    plt.plot(max_index-1,scores[max_index],'D',c=col[count])
    count+=1
    
plt.legend(frameon=False,loc=(0,1))
plt.ylabel('Score')
plt.xlabel('Iteration Number')
plt.ylim(5,15)
plt.xlim(0,400)
plt.savefig('timeseries.png',dpi=600)
```
